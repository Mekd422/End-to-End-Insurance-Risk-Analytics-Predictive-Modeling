```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("ggplot")

df = pd.read_csv("../data/interim/cleaned.csv", sep=",", low_memory=False)

df.head()
```

```python
print("Shape:", df.shape)
df.info()
```

```python
df.describe().T
```

```python
df.describe(include="object").T
```

```python
missing = df.isnull().mean().sort_values(ascending=False)
missing.head(20)

```

```python
missing = df.isnull().mean().sort_values(ascending=False)
missing.head(20)
```

```python
plt.figure(figsize=(10,6))
missing.head(20).plot(kind='barh')
plt.title("Top Missing Values")
plt.show()
```

```python
df['lossratio'] = df['TotalClaims'] / df['TotalPremium']

portfolio_lr = df['lossratio'].mean()
portfolio_lr
```

```python
lr_province = df.groupby("Province")["lossratio"].mean().sort_values()
lr_vehicle = df.groupby("VehicleType")["lossratio"].mean().sort_values()
lr_gender = df.groupby("Gender")["lossratio"].mean().sort_values()

lr_province, lr_vehicle, lr_gender
```

```python
df['transactionmonth'] = pd.to_datetime(df['TransactionMonth'])

monthly = df.groupby(df.transactionmonth.dt.to_period("M"))[
    ["TotalPremium", "TotalClaims"]
].sum()

monthly.index = monthly.index.to_timestamp()

plt.figure(figsize=(12,5))
plt.plot(monthly.index, monthly['TotalPremium'], label='Total Premium')
plt.plot(monthly.index, monthly['TotalClaims'], label='Total Claims')
plt.legend()
plt.title("Monthly Premium vs Claim Trends")
plt.show()
```

```python
claims_by_make = df.groupby("make")["TotalClaims"].sum().sort_values(ascending=False)

claims_by_make.head(10)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols].corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['TotalClaims'])
plt.title("Outliers: TotalClaims")
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(x=df['CustomValueEstimate'])
plt.title("Outliers: CustomValueEstimate")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=lr_province.index, y=lr_province.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Loss Ratio by Province")
plt.ylabel("Loss Ratio")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.regplot(x="TotalPremium", y="TotalClaims", data=df, scatter_kws={'alpha':0.3})
plt.title("Premium vs Claims")
plt.show()
```

```python
top_makes = claims_by_make.head(12)

plt.figure(figsize=(12,6))
sns.barplot(x=top_makes.index, y=top_makes.values, palette="magma")
plt.xticks(rotation=45)
plt.title("Top 12 Vehicle Makes by Total Claims")
plt.ylabel("Total Claims")
plt.show()
```
